```python

import matplotlib.pyplot as plt
import numpy as np
import cv2

filename = 'noise_img.jpg'
image = np.array(cv2.imread(filename))
nor_img = np.zeros(shape = (image.shape[0]*2, image.shape[1]*2))
image = cv2.normalize(image, nor_img, 0, 255, cv2.NORM_MINMAX)
plt.imshow(image)
```
